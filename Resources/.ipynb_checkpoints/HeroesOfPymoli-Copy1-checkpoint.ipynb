```python
# Dependencies and Setup
import pandas as pd

# File to Load 
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#display all columns and rows in csv
players_df = pd.DataFrame(purchase_data)
players_df.head()
```

## Player Count

* Display the total number of players


```python
#find total players by unique screen name
players_unique_count = players_df["SN"].nunique()

#create dataframe with total players
total_players_df = pd.DataFrame({"Total Players":[players_unique_count]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#sum, average, convert days to week, 2:10 
#output to csv 2:41
#replace names 1:03 - day 2

# unique items
unique_items = players_df["Item ID"].nunique()

#average_price
average_price = players_df["Price"].mean()

#total purchases
total_purchases = players_df["Purchase ID"].count()

#total revenue
total_revenue = players_df["Price"].sum()

#create dataframe
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [average_price], 
                                       "Number of Purchases": [total_purchases], "Total Revenue": [total_revenue]})

#format for currency
purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:.2f}".format)

#summary
purchasing_analysis_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# group by gender 
gender_stats = players_df.groupby(["Gender"])
gender_stats.count()

#total purchases by gender
purchase_count = gender_stats["Purchase ID"].count()

#avg purchase price by gender
avg_purchase_price = gender_stats["Price"].mean()

#total purchase value by gender
total_purchase_value = gender_stats["Price"].sum()

#avg purchase total per person
avg_purchase_total_pp = (avg_purchase_value / purchase_count)

# #summary
purchasing_analysis_gender_df = pd.DataFrame({"Purchase Count": purchase_count,"Purchase Price": avg_purchase_price,
                                              "Total Purchase Value": total_purchase_value, "Avg Total Purchase per Person": avg_purchase_total_pp})

#format table
purchasing_analysis_gender_df["Purchase Price"] = purchasing_analysis_gender_df["Purchase Price"].map("${:.2f}".format)
purchasing_analysis_gender_df["Total Purchase Value"] = purchasing_analysis_gender_df["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis_gender_df["Avg Total Purchase per Person"] = purchasing_analysis_gender_df["Avg Total Purchase per Person"].map("${:.2f}".format)

#display table
purchasing_analysis_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Bins for ages   - 1:13
age_bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 100]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#sort different ages into bin
players_df["Age Group"] = pd.cut(players_df["Age"], age_bins, labels=age_labels, include_lowest=True)

#group by age - 1:46 class 3
age_group = players_df.groupby("Age Group")
age_group.count()

#Total count by age
total_count_age = age_group["SN"].nunique()

#Percentage
percent_age = (total_count_age / total_players) * 100

#Summary
age_demo_df = pd.DataFrame({"Total Count": total_count_age, "Percentage of Players": percent_age})
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#total purchases by age group
total_purchase_age = age_group["Purchase ID"].nunique()

#average purchase price by age 
avg_purchase = age_group["Price"].mean()
avg_purchase

#total purchase value
total_purchase_value = age_group["Price"].sum()
total_purchase_value

#avg total purchase per person
avg_total_pp = total_purchase_value / total_count_age
avg_total_pp

#summary
purch_df = pd.DataFrame({"Purchase Count": total_purchase_age, "Average Purchase Price": avg_purchase, 
                         "Total Purchase Value": total_purchase_value, "Avg Total Purchase per Person": avg_total_pp})

#format dataframe
purch_df["Average Purchase Price"] = purch_df["Average Purchase Price"].map("${:.2f}".format)
purch_df["Total Purchase Value"] = purch_df["Total Purchase Value"].map("${:.2f}".format)
purch_df["Avg Total Purchase per Person"] = purch_df["Avg Total Purchase per Person"].map("${:.2f}".format)

#display dataframe
purch_df
```
